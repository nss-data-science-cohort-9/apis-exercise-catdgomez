In this exercise, you'll redo the data gathering phase of the UNData Exploration project by using APIs instead of downloading csv files.

#### 1. Use the API to get all available data for the GDP per capita, PPP (constant 2021 international $) indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
url_input = 'NY.GDP.PCAP.PP.KD'

params = {
    'format' : 'json',
    'per_page' : '20000'
}

endpoint = f'https://api.worldbank.org/v2/country/all/indicator/{url_input}'

response = requests.get(endpoint, params=params)
resp = response.json()
gdp_df = pd.json_normalize(resp[1])
gdp_df.head(2)
gdp_df
```

```python
codes = 'NY.GDP.PCAP.PP.KD'
country_input = 'usa'
params = {
    'format' : 'json',
    'per_page' : '20000'
}

def data_pull(*args) :
    endpoint = f'https://api.worldbank.org/v2/country/{country_input}/indicator/{codes}'
    response = requests.get(endpoint, params=params)
    resp = response.json()
    gdp_df = pd.json_normalize(resp[1])
    return gdp_df 

data_pull(url_input, country_input).head(2)
```

#### 3. Adjust your request so that it returns data just for the United States for the years 2000 through 2021.



```python
range_ending = 'SP.DYN.LE00.IN'
country_call = 'usa'
start_date=2000 
end_date=2021
params = {
    'format' : 'json',
    'per_page' : '20000'
}

def country_date_range(*args):#, year=None) :
    if year == None :
        endpoint = f'https://api.worldbank.org/v2/country/{country_call}/indicator/{range_ending}'
    else :
        date = f'?date={start_date}:{end_date}'
        endpoint = f'https://api.worldbank.org/v2/country/{country_call}/indicator/{range_ending}{date}'
    response = requests.get(endpoint, params=params)
    resp = response.json()
    res = pd.json_normalize(resp[1])
    return res 

country_date_range().head(2)
```
